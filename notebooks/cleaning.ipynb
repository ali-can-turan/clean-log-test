# Clean
## Initialize

Column names will be freed of unwanted char '@':

```python
df.columns = [column.replace("@", "_") for column in df.columns.tolist()]
```

An object from Cleaning class will be generated below to handle some of the cleaning operations while carrying the self-object 'df'. 

```python
cleaner = cleaning.Cleaning(df=df)
```

## Initial cleaning

Columns' content will be freed from unwanted chars below via Cleaner method 'str_clean':

```python
cleaner.str_clean(columns_chars= {"Contact_Info": r".*?[:_]", "Order_ID":r".*?[:_]", "Customer_Name": r".*?\."}, regex=True).head()
```

Columns' content will be assigned cases as desired below via Cleaner method 'assign_cases':

```python
cleaner.case_assign(columns_cases={"Product_Category": "title", "Customer_Name": "title"}).head()
```

## Formatting

Columns' content will be formatted as desired below:

```python
cleaner.df = cleaner.df.astype({
	"Order_ID" : "string",
	"Customer_Name" : "string",
	"Contact_Info" : "string",
	"Product_Category" : "string",
	"Regional_Code" : "string",
	"Sale_Price": "float64",
	"Inventory_Count": "float64"
})
cleaner.df.info()
```

```python
cleaner.df["Sale_Price"] = cleaner.df["Sale_Price"].round(2)
cleaner.df["Inventory_Count"] = cleaner.df["Inventory_Count"].round().astype("Int64")
```

Date columns' data type will be changed as desired below, via Cleaner method 'datetime_converter', to be able to conduct calculations if needed down the run:

```python
cleaner.datetime_converter(columns=["Order_Date", "Delivery_Date"], to="datetime").info()
```

## Missing values

Missing values in the columns will be observed below via constructing desired boolean relations between columns through Cleaner method 'null_rows' to decide in the next step how to treat them:

```python
columns= ["Customer_Name", "Contact_Info"]
cleaner.null_rows(columns=columns, separator="&")

columns= ["Order_ID", "Sale_Price", "Order_Date", "Delivery_Date"]
cleaner.null_rows(columns=columns, separator="|")
```

The rows observed above with null values will be dropped below:

```python
cleaner.df = cleaner.df.dropna(how="all", subset=["Customer_Name", "Contact_Info"], axis=0).reset_index(drop=True)
cleaner.df = cleaner.df.dropna(how="any", subset=["Order_ID", "Sale_Price", "Order_Date", "Delivery_Date"], axis=0).reset_index(drop=True)
```

## Duplicates

Total row count that contains duplicated values in the "Order_ID" column will be observed as no duplication in this column is envisaged:

```python
cleaner.df["Order_ID"].duplicated().sum()
```

```python
cleaner.df.duplicated().sum()
```

## Outliers

Below, discount_Rate column values are capped at the value 0.75 as no further discount rate is assumed:

```python
cleaner.df.loc[:, "Discount_Rate"] = cleaner.df["Discount_Rate"].clip(upper=0.75)
cleaner.df[["Discount_Rate"]].sort_values(by="Discount_Rate", ascending=False).head()
```

Below z_score method will be employed to track the outliers in the Sale_Price column via Cleaner method 'outlier_z'. It will show the range bins of z_score values:

```python
cleaner.outlier_z(op="bins", col="Sale_Price", group_by="Product_Category", z_abs_thresh=1)
```

Below iqr method will be employed to track the outliers in the Sale_Price column, through Cleaner method 'outlier_iqr'. It will show the range bins of outliers:

```python
cleaner.outlier_iqr(op="bins", col="Sale_Price", coef=0.5)
```

Below df_z will be created out of the z_score values in the df:

```python
df_z = cleaner.outlier_z(op="adopt", col="Sale_Price", group_by="Product_Category", z_abs_thresh=1)
df_z
```

Below df_iqr will be created out of the iqr outliers in the df:

```python
df_iqr = cleaner.outlier_iqr(op="adopt", col="Sale_Price", coef=0.5)
df_iqr
```

Below df_z will be selected and assigned to the df that the cleaner object carries:

```python
cleaner.df = df_iqr
```

# Transform
## Manipulation

Below, columns will be reordered as desired, through Cleaner method 'column_reorder':

```python
cleaner.column_reorder(columns_idx={"Customer_Name": 1, "Inventory_Count": 4, "Discount_Rate": 6, 'z_scores_Sale_Price': 7}).columns.tolist()
```

## Enrich

Below, two columns -'Email' and 'Tel_Nu'- will be generated and populated with the content of 'Cantact_Info' column as this column both incoporates email and telephone number values at the same time: 

```python
cleaner.df.loc[:, "Email"] = df["Contact_Info"].where(df["Contact_Info"].str.contains("com|org", na=False)).astype("string")
cleaner.df.loc[:, "Tel_Nu"] = df["Contact_Info"].where(~df["Contact_Info"].str.contains("com|org", na=False)).astype("string")
cleaner.df[["Email", "Tel_Nu"]].head()
cleaner.df = cleaner.df.drop(columns=("Contact_Info"))
```

Below, columns will be reordered as desired, through Cleaner method 'column_reorder':

```python
cleaner.column_reorder({"Email": 2, "Tel_Nu": 2}).columns.tolist()
```

Below, new row entries will be generated through Cleaner method 'numeric_entry_df_generator' -with only numeric columns populated- for only to be incoporated into the df  / cleaner object later on:

```python
df_new_entry = cleaner.numeric_entry_df_generator(entry_num=3)
df_new_entry
```

Below: new row entries will be concatenated into the df:

```python
cleaner.df = pd.concat([cleaner.df, df_new_entry], ignore_index=True, axis=0)
cleaner.df
```

# Validate

Below, it will be validated that the all regional codes in the 'Regional_Code' column starts with the defined region abbreviations: 

```python
regions = ["USA", "CAN", "GER", "UK"]
mask_regions = cleaner.df["Regional_Code"].str.split("-").str[0].isin(regions)
mask_na = cleaner.df["Regional_Code"].isna()
mask_regions.sum() + mask_na.sum() == cleaner.df.shape[0]
```

Below, it will be validated that no date value in the Order_Date column is more recent than the corresponding date value in the 'Delivery_Date' column:

```python
any(cleaner.df['Order_Date'] > cleaner.df['Delivery_Date'])
```

# Operations and Aggregation

Below, product categories will be observed to be able to conduct groupings over the 'Product_Category' column later on:

```python
set(df["Product_Category"])
```

Below: two new columns -'Avg_Sale_Price_By_Product_Category' and 'Total_Inventory_Count_By_Product_Category'- will be generated via aggregations over 'Sale_Price' and 'Inventory_Count' columns respectively by grouping product categories, to see relevant performance data by product group:

```python
cleaner.df["Avg_Sale_Price_By_Product_Category"] = cleaner.df.groupby("Product_Category")["Sale_Price"].transform(lambda cell: cell.mean().round(2))

cleaner.df["Total_Inventory_Count_By_Product_Category"] = cleaner.df.groupby("Product_Category")["Inventory_Count"].transform(func="sum")

cleaner.df[["Product_Category", "Avg_Sale_Price_By_Product_Category", "Total_Inventory_Count_By_Product_Category"]].head()
```

Below, columns will be reordered as desired via Cleaner method 'column_reorder':

```python
cleaner.column_reorder(columns_idx={'Total_Inventory_Count_By_Product_Category': 6, 'Avg_Sale_Price_By_Product_Category': 8})
```

Below: 'Regional_Code' is grouped and 'Inventory_Count' and 'Order_ID' columns will be aggregated via mean and count functions respectively to see the products of which reginal code have what inventory count and brought about how many orders:

```python
df_regional_code = cleaner.df.groupby("Regional_Code").agg({"Inventory_Count": "mean", "Order_ID": "count"})
df_regional_code
```

Below, the df will be resampled on 'Order_Date' date column to have a holistic / monthly insight on the numeric columns:

```python
cleaner.df.info()
df_resampled = cleaner.df.resample(on="Order_Date", rule="W", closed="left", label="left").mean(numeric_only=True).drop(columns="z_scores_Sale_Price")
df_resampled
```
